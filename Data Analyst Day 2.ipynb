```python
# String is data structure in python that represent sequence of character,it immuteable it means once you created it can not changed again
# each character is encoded inside the assci or unicode character,so we can say strings is collection of assci or unicode characters
a='Hello students how are you'
print(a,type(a))
# indexing in string Hello left to right(0,1,2,3,4),right to left negative indexing(-1,-2,-3,-4,-5)
print(a[0])
print(a[-1])
# slicing of string means we create sub string of string
# slicing syntax [start:end:steps](if you want to step give 3 and step 1 give 2 do like this)
print(a[0:5])
print(a[0:12:3])
# reverse indexing 
print(a[::-1])
# reverse slicing
print(a[-3:-1])
# multiline string
a1="""  Pakistan,[e] officially the Islamic Republic of Pakistan,[f] is a country in South Asia. It is the fifth-most populous country, with a population of over 241.5 million, having the second-largest Muslim population as of 2023.[8] Islamabad is the nation's capital, while Karachi is its largest city and financial centre. Pakistan is the 33rd-largest country by area and the ninth-largest in Asia. Bounded by the Arabian Sea on the south, the Gulf of Oman on the southwest, and the Sir Creek on the southeast, it shares land borders with India to the east; Afghanistan to the west; Iran to the southwest; and China to the northeast. It shares a maritime border with Oman in the Gulf of Oman, and is separated from Tajikistan in the northwest by Afghanistan's narrow Wakhan Corridor.  """
print(a1)
# find the length of string
print(len(a))
# convert whole string into upper case,use upper method
print(a.upper())
# convert whole string into upper case,use upper method
a1='LOWER CONVERSION OF STRING'
print(a1.lower())
# we know that string is immuteable we canot change it but we use replace function that creates the copy of string and make change in it
# replace function is used to make change in the string
print(a1.replace('LOWER','By'))
# find method is used to find the position or index of string character
print(a1.find('L'))
```

```python
# list is used to store multiple items in a single variable like tuple,dict,set but it is mutable it can be altered after it creation
# single list may contain diferent data types items like int,string,float etc,
# it is ordered or indexed we can access its items by its index left to right(0,1,2,3,4),right to left negative indexing(-1,-2,-3,-4,-5)
# lists allows duplicates
l1=[1,2,3,4,5,6,7,'string',['a','b','c'],(123,321),1]
print(l1,type(l1))
print(l1[9])
print(l1[8])
print(l1[-9])
# len function used to find the length of list
print(len(l1))
# slicing of list
print(l1[2:5:2])
# change items in the list
l1[1]='strawberry'
print(l1)
# change the range of items
l2=[1,2,3,4,5,6,7]
l2[2:4]=['pen','pencil','copy']
print(l2)
# append method is used to add the value in the list at the end
l2.append('Laptop')
print(l2)
# remove mehod is used to remove the item in the lis
l2.remove('pen')
print(l2)
# reverse function is used to reverse the list
l2.reverse()
print(l2)
```

```python
# tuple is also use to store multiple items in the single variable 
# it is simillar to the list like indexing or ordered ,duplication of values,and we can store different data types data
# but the differece between list and tuple is ,tuple is immuteable we canot change its value once create
t=(1,'abc',3.1,(43,56,78),['bdo',12])
print(t)
print(t[2])
# if we want to store 1 item in tuple we must use , after the item
t1=(456,)
print(t1,type(t1))
# we can create tuple and list by tuple and list constructor
lc1=list([23,45,56,78])
print(lc1,type(lc1))
tc=tuple((123,'acv',))
print(tc,type(tc))
# indexing left to right(0,1,2,3,4),right to left negative indexing(-1,-2,-3,-4,-5)
print(t[-1])
# negative slicing
print(t[-3:-1])
# positive slicing
print(t[1:3])
# append in the tuple
lc1.append(30)
print(lc1)
# remove item from the tuple
lc1.remove(78)
print(lc1)
# del with tuple
tu=(11,23,45)
print(tu)
del tu #it will delete the tuple from the memory
# print(tu)
# join  two tuples
ta=(1,2,3,4)
tb=(5,6,7,8)
print(ta+tb)
# multiply tuple
print(ta*3)
# add tuple
print(ta+ta+ta)

```

```python
# dictionaries are used to store data values in key:value pair 
# ict an list share the folowing characterstics 
# 1 both are mutable
# 2 both can be nested
# dict items are ordered ,changable,and does not allow duplicate values 
d={"id":"123",
   "name":"abc",
   "class":16,
   "l1":[1,2,3,4,5,6],
   3:{1:2,3:4},
   'd2':{'a':'12','b':'cd'}
  
}
print(d['l1'],d['d2'])
# creating dictionary by constructor
di=dict(name='abc',age=23)
print(di,type(di))
# can reterive value of key by get function
print(d.get('id'))
# can see all keys
print(d.keys())
# can see all values
print(d.values())
# we can change the value of pair because it is mutable,or can change item value
print(d)
d['id']=34
print(d)
# add new item or key value pair in dict
d.update({5:'none'})
print(d)
# we can delete item or key value pair by pop function
d.pop(5)
print(d)
# we can delete the key value pair or dict by del method
del d['class']
print(d)
d34={1:'we'}
print(d34)
del d34
# print(d34) now the d34 is deleted
# clear method delete all the items from the dict
d33={1:11,2:22}
print(d33)
d33.clear()
print(d33)

```
